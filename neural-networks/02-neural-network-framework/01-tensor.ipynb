# Tensor

```python
import numpy as np
```

## Foundation

### Tensor

```python
class Tensor:

    def __init__(self, data):
        self.data = np.array(data)

    def size(self):
        return len(self.data)

    def __str__(self):
        return str(self.data)
```

## Data

### Features

```python
feature = Tensor([28.1, 58.0])
```

## Model

### Weight and Bias

```python
weight = Tensor(np.ones([1, 2]) / 2)
bias = Tensor(np.zeros(1))
```

### Prediction Function

```python
def forward(x, w, b):
    return Tensor(x.data @ w.data.T + b.data)
```

## Testing

### Predicting

```python
prediction = forward(feature,
                     weight,
                     bias)

print(f'prediction: {prediction}')
```
