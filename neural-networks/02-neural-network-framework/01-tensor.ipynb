```python
import numpy as np
```

## åŸºç¡€æ¶æ„

### å¼ é‡

æˆ‘ä»¬é¦–å…ˆå¼•å…¥çš„ç»“æ„ï¼Œæˆ–è€…ç±»ï¼Œæ˜¯**å¼ é‡**ï¼ˆTensorï¼‰ã€‚

å¼ é‡é¦–å…ˆæ˜¯å¯¹ NumPy æ•°ç»„çš„åŒ…è£…ã€‚åœ¨å¼ é‡å†…éƒ¨ï¼Œæˆ‘ä»¬å°†ç»§ç»­ä½¿ç”¨ NumPy æ•°ç»„å®Œæˆæ•°æ®ä¿å­˜å’Œè®¡ç®—ã€‚

åç»­ï¼Œæˆ‘ä»¬å°†ä»¥å¼ é‡ä¸ºåŸºç¡€ï¼Œé€æ­¥å®ç°**è‡ªåŠ¨å¾®åˆ†**ï¼ˆAutogradï¼‰çš„æœºåˆ¶ã€‚

å¼ é‡å°†æˆä¸º**æ¢¯åº¦è®¡ç®—é“¾è·¯**çš„ä¸»è¦è½½ä½“ã€‚

```python
class Tensor:

    def __init__(self, data):
        self.data = np.array(data)

    def __repr__(self):
        return f'Tensor({self.data})'
```

## æ•°æ®

### ç‰¹å¾

åœ¨æ–°çš„æ¡†æ¶é‡Œï¼Œæ‰€æœ‰æ ·æœ¬éƒ½å°†ä»¥å¼ é‡çš„å½¢å¼æä¾›ã€‚

```python
feature = Tensor([28.1, 58.0])
```

## æ¨¡å‹

### å‚æ•°ï¼šæƒé‡ã€åç½®

æ‰€æœ‰æ¨¡å‹å‚æ•°ä¹Ÿéƒ½å°†ä»¥å¼ é‡çš„å½¢å¼æä¾›ã€‚

```python
weight = Tensor(np.ones((1, 2)) / 2)
bias = Tensor(np.zeros(1))
```

### æ¨ç†å‡½æ•°ï¼ˆå‰å‘ä¼ æ’­ï¼‰

æ‰€æœ‰çš„ä¸­é—´æ•°æ®ä¹Ÿéƒ½å°†ä»¥å¼ é‡çš„å½¢å¼æä¾›ã€‚

è¿™æ ·åšçš„ç›®çš„ï¼Œæ˜¯ä¸ºäº†åç»­å¯ä»¥åˆ©ç”¨å¼ é‡è‡ªåŠ¨å¾®åˆ†çš„èƒ½åŠ›ï¼Œå®ç°æ¢¯åº¦è®¡ç®—é“¾è·¯çš„è‡ªåŠ¨åŒ–ã€‚

```python
def forward(x, w, b):
    return Tensor(x.data @ w.data.T + b.data)
```

## éªŒè¯

### æ¨ç†

```python
prediction = forward(feature, weight, bias)
print(f'prediction: {prediction}')
```
