# 学习率

```python
import numpy as np
```

## Data

### Features and Label

```python
feature = np.array([28.1, 58.0])
label = np.array([165])
```

## Model

### Weight and Bias

```python
weight = np.ones([1, 2]) / 2
bias = np.zeros(1)
```

### Prediction Function

```python
def forward(x, w, b):
    return x @ w.T + b
```

### Mean Squared Error Loss Function

```python
def mse_loss(p, y):
    return ((p - y) ** 2).mean()
```

### Gradient Function

```python
def gradient(p, y):
    return (p - y) * 2
```

### Learning Rate

```python
LEARNING_RATE = 0.00001
```

### Backward Function

```python
def backward(x, d, w, b):
    w -= d * x * LEARNING_RATE
    b -= np.sum(d) * LEARNING_RATE
    return w, b
```

## Testing

### Predicting

```python
prediction = forward(feature,
                     weight,
                     bias)

print(f'prediction: {prediction}')
```

### Calculating Loss

```python
error = mse_loss(prediction,
                 label)

print(f'error: {error}')
```

### Calculating Gradient and Updating Weight and Bias

```python
delta = gradient(prediction,
                 label)
weight, bias = backward(feature,
                        delta,
                        weight,
                        bias)

print(f"weight: {weight}")
print(f"bias: {bias}")
```
