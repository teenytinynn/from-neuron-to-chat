# 迭代

```python
import numpy as np
```

## Dataset

### Training Features and Labels

```python
train_features = np.array([[22.5, 72.0],
                           [31.4, 45.0],
                           [19.8, 85.0],
                           [27.6, 63]])

train_labels = np.array([[95],
                        [210],
                        [70],
                        [155]])
```

### Testing Features and Labels

```python
test_features = np.array([[28.1, 58.0]])
test_labels = np.array([[165]])
```

## Model

### Weight and Bias

```python
weight = np.ones([1, 2]) / 2
bias = np.zeros(1)
```

### Prediction Function

```python
def forward(x, w, b):
    return x @ w.T + b
```

### Mean Squared Error Loss Function

```python
def mse_loss(p, y):
    return ((p - y) ** 2).mean()
```

### Gradient Function

```python
def gradient(p, y):
    return (p - y) * 2
```

### Learning Rate

```python
LEARNING_RATE = 0.00001
```

### Backward Function

```python
def backward(x, d, w, b):
    w -= d * x * LEARNING_RATE
    b -= np.sum(d) * LEARNING_RATE
    return w, b
```

## Training

### Iteration Training

```python
for i in range(len(train_features)):

    feature = train_features[i]
    label = train_labels[i]

    prediction = forward(feature,
                         weight,
                         bias)
    error = mse_loss(prediction,
                     label)

    delta = gradient(prediction,
                     label)
    weight, bias = backward(feature, delta,
                            weight,
                            bias)

print(f"weight: {weight}")
print(f"bias: {bias}")
```

## Testing

### Predicting

```python
predictions = forward(test_features,
                      weight,
                      bias)

print(f'predictions: {predictions}')
```

### Calculating Loss

```python
error = mse_loss(predictions,
                 test_labels)

print(f'error: {error}')
```
