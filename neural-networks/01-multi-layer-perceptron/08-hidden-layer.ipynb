# 隐藏层

```python
import numpy as np

np.random.seed(99)
```

## Dataset

### Training Features and Labels

```python
train_features = np.array([[22.5, 72.0],
                           [31.4, 45.0],
                           [19.8, 85.0],
                           [27.6, 63]])

train_labels = np.array([[95],
                        [210],
                        [70],
                        [155]])
```

### Testing Features and Labels

```python
test_features = np.array([[28.1, 58.0]])
test_labels = np.array([[165]])
```

## Model

### Hidden Layer Weights and Biases

```python
hidden_weight = np.random.rand(4, 2) / 2
hidden_bias = np.zeros(4)
```

### Output Layer Weights and Biases

```python
output_weight = np.random.rand(1, 4) / 4
output_bias = np.zeros(1)
```

### Prediction Function

```python
def forward(x, w, b):
    return x @ w.T + b
```

### Mean Squared Error Loss Function

```python
def mse_loss(p, y):
    return ((p - y) ** 2).mean()
```

### Gradient Function

```python
def gradient(p, y):
    return (p - y) * 2 / len(y)
```

### Learning Rate

```python
LEARNING_RATE = 0.00001
```

### Backward Function

```python
def backward(x, d, w, b):
    w -= d.T @ x * LEARNING_RATE
    b -= np.sum(d, axis=0) * LEARNING_RATE
    return w, b
```

### Gradient Backward Function

```python
def gradient_backward(d, w):
    return d @ w
```

## Training

### Epoch

```python
EPOCHS = 1000
```

### Batch

```python
BATCH_SIZE = 2
```

### Iterative Training

```python
for epoch in range(EPOCHS):
    for i in range(0, len(train_features), BATCH_SIZE):

        features = train_features[i: i + BATCH_SIZE]
        labels = train_labels[i: i + BATCH_SIZE]

        hidden = forward(features,
                         hidden_weight,
                         hidden_bias)
        predictions = forward(hidden,
                              output_weight,
                              output_bias)
        error = mse_loss(predictions,
                         labels)

        output_delta = gradient(predictions,
                                labels)
        hidden_delta = gradient_backward(output_delta,
                                         output_weight)

        output_weight, output_bias = backward(hidden,
                                              output_delta,
                                              output_weight,
                                              output_bias)
        hidden_weight, hidden_bias = backward(features,
                                              hidden_delta,
                                              hidden_weight,
                                              hidden_bias)

print(f"hidden weight: {hidden_weight}")
print(f"hidden bias: {hidden_bias}")
print(f"output weight: {output_weight}")
print(f"output bias: {output_bias}")
```

## Testing

### Predicting

```python
hidden = forward(test_features,
                 hidden_weight,
                 hidden_bias)
predictions = forward(hidden,
                      output_weight,
                      output_bias)

print(f'predictions: {predictions}')
```

### Calculating Loss

```python
error = mse_loss(predictions,
                 test_labels)

print(f'error: {error}')
```
