## å‡æ–¹è¯¯å·®ï¼ˆMSEï¼‰

å½“é¢å¯¹å¤šä¸ªæ ·æœ¬ï¼ˆä¾‹å¦‚ $n$ å¤©çš„é”€å”®æ•°æ®ï¼‰æ—¶ï¼Œæˆ‘ä»¬éœ€è¦ä¸€ä¸ªæ•´ä½“æŒ‡æ ‡æ¥è¡¡é‡ç½‘ç»œæ¨¡å‹çš„æ€§èƒ½ã€‚é‚£ä¹ˆå¯ä»¥è®¡ç®—æ‰€æœ‰é¢„æµ‹å€¼çš„å¹³æ–¹è¯¯å·®çš„å¹³å‡å€¼ï¼Œç§°ä¸º**å‡æ–¹è¯¯å·®**ï¼ˆMean Squared Errorï¼‰ï¼š

$$
\text{mean squared error} = \frac{1}{n} \sum\limits_{i=1}^{n} (y_i - p_i)^2
$$

è¿™é‡Œï¼š
* $y = [y_1, y_2, ..., y_n]$ï¼š$n$ ä¸ªçœŸå®ç»“æœï¼›
* $p = [p_1, p_2, ..., p_n]$ï¼š$n$ ä¸ªé¢„æµ‹ç»“æœã€‚

```python
import numpy as np
```

## æ•°æ®

ç°åœ¨ï¼Œè®©æˆ‘ä»¬å›åˆ°å°æ˜çš„å†°æ¿€å‡Œåº—ã€‚

å½“å¤©æ™šä¸Šï¼Œé”€å”®æ•°æ®å‡ºç‚‰ï¼šä¸€å…±å–å‡ºäº† 165 ä¸ªå†°æ¿€å‡Œã€‚

### ç‰¹å¾ã€æ ‡ç­¾

æˆ‘ä»¬æŠŠçœŸå®ç»“æœç§°ä¸º**æ ‡ç­¾å€¼**ï¼ˆLabelï¼‰ï¼ŒåŒæ ·ç”¨ NumPy çš„æ•°ç»„æ¥ä¿å­˜ã€‚

```python
feature = np.array([28.1, 58.0])
label = np.array([165])
```

## æ¨¡å‹

### å‚æ•°ï¼šæƒé‡ã€åç½®

```python
weight = np.ones((1, 2)) / 2
bias = np.zeros(1)
```

### æ¨ç†å‡½æ•°

```python
def forward(x, w, b):
    return x @ w.T + b
```

### æŸå¤±å‡½æ•°ï¼ˆå‡æ–¹è¯¯å·®ï¼‰

æŸå¤±å‡½æ•°ç”¨äºè®¡ç®—é¢„æµ‹å€¼ $p$ å’Œæ ‡ç­¾å€¼ $y$ çš„å‡æ–¹è¯¯å·®ã€‚å…¶ç»“æœè¢«ç§°ä¸º**æŸå¤±å€¼**ï¼ˆLossï¼‰ã€‚

```python
def mse_loss(p, y):
    return np.mean(np.square(y - p))
```

## éªŒè¯

### æ¨ç†

```python
prediction = forward(feature, weight, bias)
print(f'prediction: {prediction}')
```

### è¯„ä¼°

æœ‰äº†æŸå¤±å‡½æ•°ï¼Œæˆ‘ä»¬å°±å¯ä»¥é‡åŒ–æ¨¡å‹é¢„æµ‹çš„è¯¯å·®ã€‚

```python
loss = mse_loss(prediction, label)
print(f'loss: {loss}')
```
