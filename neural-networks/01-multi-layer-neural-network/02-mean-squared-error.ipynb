## å¹³å‡å¹³æ–¹å·®ï¼ˆMSEï¼‰

æ›´è¿›ä¸€æ­¥ï¼Œå¦‚æœç»“æœä¸æ˜¯ä¸€ä¸ªæ•°å€¼ï¼Œè€Œæ˜¯ä¸€ç»„nä¸ªæ•°å€¼ã€‚é‚£ä¹ˆå¯ä»¥è®¡ç®—æ¯å¯¹æ•°å€¼çš„å¹³æ–¹å·®ï¼Œå¹¶æœ€ç»ˆæ¯”è¾ƒæ‰€æœ‰å¹³æ–¹å·®çš„å¹³å‡å€¼ï¼Œç§°ä¸º**å¹³å‡å¹³æ–¹å·®**ï¼ˆMean Squared Errorï¼‰ï¼š

$$
\text{mean squared error} = \frac{1}{n} \sum\limits_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

è¿™é‡Œï¼š
* å®é™…ç»“æœï¼š$y = [y_1, y_2, ..., y_n]$
* æ¨ç†ç»“æœï¼š$\hat{y} = [\hat{y}_1, \hat{y}_2, ..., \hat{y}_n]$

```python
import numpy as np
```

## æ•°æ®

ç°åœ¨ï¼Œè®©æˆ‘ä»¬å›åˆ°å°æ˜çš„å†°æ¿€å‡Œåº—ã€‚

å½“å¤©æ™šä¸Šï¼Œé”€å”®é¢ç»Ÿè®¡å‡ºæ¥äº†ã€‚ä»Šå¤©ä¸€å…±å–å‡ºäº†165ä¸ªå†°æ¿€å‡Œã€‚

### ç‰¹å¾ã€æ ‡ç­¾

æˆ‘ä»¬æŠŠå®é™…ç»“æœç§°ä¸º**æ ‡ç­¾å€¼**ï¼ˆLabelï¼‰ï¼Œæˆ‘ä»¬åŒæ ·ä½¿ç”¨NumPyçš„æ•°ç»„æ¥ä¿å­˜ã€‚

```python
feature = np.array([28.1, 58.0])
label = np.array([165])
```

## æ¨¡å‹

### å‚æ•°ï¼šæƒé‡ã€åå·®

```python
weight = np.ones([1, 2]) / 2
bias = np.zeros(1)
```

### æ¨ç†å‡½æ•°

```python
def forward(x, w, b):
    return x @ w.T + b
```

### æŸå¤±å‡½æ•°ï¼ˆå¹³å‡å¹³æ–¹å·®ï¼‰

æŸå¤±å‡½æ•°ç”¨äºè®¡ç®—é¢„æµ‹å€¼å’Œæ ‡ç­¾å€¼çš„å¹³å‡å¹³æ–¹å·®ã€‚å…¶ç»“æœè¢«ç§°ä¸º**æŸå¤±å€¼**ï¼ˆLossï¼‰ã€‚

```python
def mse_loss(p, y):
    return np.mean(np.square(y - p))
```

## éªŒè¯

### æ¨ç†

```python
prediction = forward(feature, weight, bias)

print(f'prediction: {prediction}')
```

### è¯„ä¼°

æœ‰äº†æŸå¤±å‡½æ•°ï¼Œæˆ‘ä»¬å°±å¯ä»¥è¯„ä¼°äººå·¥ç¥ç»å…ƒç½‘ç»œæ¨¡å‹æ¨ç†çš„è¯¯å·®å¤§å°ã€‚

```python
error = mse_loss(prediction, label)

print(f'loss: {error}')
```
